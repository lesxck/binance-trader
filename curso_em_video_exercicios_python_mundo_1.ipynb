## O que é Type Casting (ou Conversão de Tipo)?

**Type Casting** (ou **Conversão de Tipo**) é o processo de mudar o tipo de dado de uma variável de um tipo para outro.

Em Python, como vimos no exemplo anterior, a função `input()` sempre retorna uma `string`. No entanto, muitas vezes precisamos trabalhar com números inteiros (`int`) ou decimais (`float`). É aí que o type casting se torna essencial.

### Exemplos de Type Casting:

*   **`int()`**: Converte um valor para um número inteiro.
    ```python
    texto_numero = "10"
    numero_inteiro = int(texto_numero) # numero_inteiro será 10 (tipo int)
    ```

*   **`float()`**: Converte um valor para um número decimal (ponto flutuante).
    ```python
    texto_decimal = "10.5"
    numero_decimal = float(texto_decimal) # numero_decimal será 10.5 (tipo float)
    ```

*   **`str()`**: Converte um valor para uma string (texto).
    ```python
    valor_inteiro = 25
    texto_valor = str(valor_inteiro) # texto_valor será "25" (tipo str)
    ```

### Por que é importante?

Realizar a conversão de tipo é crucial para:
1.  **Realizar operações corretas**: Você não pode fazer cálculos matemáticos com strings, por exemplo. Ao converter para `int` ou `float`, as operações funcionam como esperado.
2.  **Garantir compatibilidade**: Algumas funções ou métodos esperam um tipo de dado específico. O type casting permite que você forneça o tipo correto.
3.  **Evitar erros**: Tentar usar um valor do tipo errado pode levar a erros de execução no seu programa.

### Por que `input()` retorna sempre uma string?

A função `input()` lê a entrada do teclado como uma sequência de caracteres. Por exemplo, se você digita `123`, para o Python, isso é a string `"123"`, e não o número inteiro `123`. Se você digita `hello`, também é a string `"hello"`.

Para que o Python trate esses caracteres como um número, um booleano ou qualquer outro tipo, você deve usar as funções de conversão de tipo (`int()`, `float()`, etc.).

### Lidando com erros na conversão

É importante notar que, se o usuário digitar algo que não pode ser convertido para o tipo desejado (por exemplo, digitar 'abc' quando se espera um `int`), seu programa pode gerar um erro (`ValueError`). Podemos usar blocos `try-except` para lidar com essas situações de forma mais robusta.

````python
print('--- Exemplo de Conversão Segura com try-except ---')

try:
    idade_str = input('Digite sua idade (número inteiro): ')
    idade = int(idade_str) # Tentando converter para inteiro
    print(f"Sua idade é: {idade}, do tipo: {type(idade)}")
except ValueError:
    print(f"Erro: '{idade_str}' não é um número inteiro válido. Por favor, digite apenas números.")

print('\n--- Outro Exemplo com float ---')

try:
    preco_str = input('Digite o preço (use ponto para decimais): ')
    preco = float(preco_str) # Tentando converter para float
    print(f"O preço digitado é: {preco}, do tipo: {type(preco)}")
except ValueError:
    print(f"Erro: '{preco_str}' não é um número decimal válido. Por favor, use o formato correto (ex: 19.99).")

print('\nIsso mostra que a conversão não é automática, mas é essencial para trabalhar com tipos diferentes de strings.')
````
