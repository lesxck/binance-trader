### Exercício 008: Escreva um programa que leia um valor em metros e o exiba convertido em centímetros e milímetros.

```python
n1 = (input('digite os metros que deseja converter: '))
n = float(n1.replace (',' , '.'))
cdc = (n * 10)
ccm = (n * 100)
cmm = (n * 1000)
cdm = (n / 10)
chm = (n / 100)
ckm = (n / 1000)
print(f'O valor de {n}M, em decimetros é: {cdc :.1f}')
print(f'O valor de {n}M, em centimetros é: {ccm :.1f}CM')
print(f'O valor de {n}M, em militros é: {cmm :.1f}MM')
print(f'O valor de {n}M, em decametro é: {cdm :.1f}DM')
print(f'O valor de {n}M, em hectometro é: {chm :.2f}HM')
print(f'O valor de {n}M, em kilometro é: {ckm :.3f}KM')
```

### Exercício 009: Faça um programa que leia um número Inteiro qualquer e mostre na tela a sua tabuada.

```python
n = int(input('digite um numero inteiro e obtenha sua tabuada: '))
print(f'{n} X 1 = {n*1 :1}')
print(f'{n} X 2 = {n*2 :1}')
print(f'{n} X 3 = {n*3 :1}')
print(f'{n} X 4 = {n*4 :1}')
print(f'{n} X 5 = {n*5 :1}')
print(f'{n} X 6 = {n*6 :1}')
print(f'{n} X 7 = {n*7 :1}')
print(f'{n} X 8 = {n*8 :1}')
print(f'{n} X 9 = {n*9 :1}')
print(f'{n} X 10= {n*10 :1}')
```
