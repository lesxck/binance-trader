### Classificador de Triângulos (completo)

Este programa solicita três comprimentos de lados e, se eles puderem formar um triângulo, classifica-o por seus lados e por seus ângulos.

```python
print('--- Classificador de Triângulos ---')

# Ler os três lados, tratando a vírgula como separador decimal
lado1_str = input('Digite o comprimento do primeiro lado: ')
lado2_str = input('Digite o comprimento do segundo lado: ')
lado3_str = input('Digite o comprimento do terceiro lado: ')

try:
    lado1 = float(lado1_str.replace(',', '.'))
    lado2 = float(lado2_str.replace(',', '.'))
    lado3 = float(lado3_str.replace(',', '.'))
except ValueError:
    print('Entrada inválida. Por favor, digite números válidos.')
else:
    print(f'Lados digitados: {lado1}, {lado2}, {lado3}')

    # 1. Verificar se os lados podem formar um triângulo
    # A soma de dois lados deve ser maior que o terceiro lado
    if not (lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1):
        print('Os valores informados NÃO podem formar um triângulo.')
    else:
        print('Os valores informados PODEM formar um triângulo.')

        # 2. Classificar o triângulo quanto aos lados
        if lado1 == lado2 == lado3:
            tipo_lados = 'EQUILÁTERO'
        elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
            tipo_lados = 'ISÓSCELES'
        else:
            tipo_lados = 'ESCALENO'

        print(f'O triângulo é {tipo_lados}.')

        # 3. Classificar o triângulo quanto aos ângulos (Teorema de Pitágoras e relações)
        # Ordenar os lados para que c seja sempre o maior (hipotenusa potencial)
        lados_ordenados = sorted([lado1, lado2, lado3])
        a, b, c = lados_ordenados[0], lados_ordenados[1], lados_ordenados[2]

        # Usar uma pequena tolerância para comparação de floats
        tolerancia = 0.001

        if abs(a**2 + b**2 - c**2) < tolerancia:
            print('Também é um triângulo RETÂNGULO.')
        elif a**2 + b**2 > c**2:
            print('Também é um triângulo ACUTÂNGULO.')
        else: # a**2 + b**2 < c**2
            print('Também é um triângulo OBTUSÂNGULO.')
```

### Cálculo dos Vértices de um Triângulo

Determinar as coordenadas dos vértices de um triângulo, tendo apenas os comprimentos dos três lados, **não é um problema com uma única solução** no plano cartesiano. Existem infinitas posições para o triângulo, dependendo da sua rotação e translação. No entanto, se fizermos algumas suposições, podemos encontrar as coordenadas:

1.  **Assumindo um Vértice na Origem e outro no Eixo X:**
    *   Podemos colocar o primeiro vértice `A` em `(0, 0)`. `
    *   Podemos colocar o segundo vértice `B` ao longo do eixo X em `(lado_c, 0)`, onde `lado_c` é um dos comprimentos dos lados.
    *   Para encontrar o terceiro vértice `C = (x, y)`, usamos as distâncias `lado_a` (distância de B a C) e `lado_b` (distância de A a C):
        *   `x² + y² = lado_b²` (distância de A a C)
        *   `(x - lado_c)² + y² = lado_a²` (distância de B a C)

    Resolvendo esse sistema de equações, você pode encontrar as coordenadas `(x, y)` para o vértice `C`.

**Exemplo Simplificado (sem código para resolver o sistema aqui):**

Se você tiver lados `a=3`, `b=4`, `c=5` (um triângulo retângulo):

*   `A = (0, 0)`
*   `B = (5, 0)` (usando `c` como a base)
*   Então, para encontrar `C = (x, y)`:
    *   `x² + y² = 4²` (lado `b`)
    *   `(x - 5)² + y² = 3²` (lado `a`)

    A solução para este caso é `C = (3, 4)` (ou `(3, -4)`).

Para um sistema genérico, é mais comum que você já tenha as coordenadas de pelo menos dois vértices, ou informações angulares, para então calcular o restante.
